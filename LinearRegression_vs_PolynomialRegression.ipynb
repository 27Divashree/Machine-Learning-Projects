Import Library

```python
#Dataset
df = pd.read_csv('MPG.csv')
```

Removing Null Values

```python
df.dropna(inplace=True)
```

Transforming categorical values to numerical values

```python
df['origin'].value_counts()
```

```python
df['origin'].unique()
```

```python
df['name'].unique()
```

```python
# replacing values
#df1 = df['origin'].replace(['usa', 'japan','europe'],[0, 1, 2], inplace=True)
df = df.replace({'origin': {'usa':1,'japan':2,'europe':3}})
```

```python
df.head()
```

Visualize Data

```python
sns.pairplot(df)
#Gives a scatter plot- gives a sense of correlation
#Here we have positive correlation- X increases -> Y increases
```

Define y and X

```python
df.columns
```

```python
y = df['mpg']
```

```python
X = df[['horsepower', 'weight']]
```

```python
#For each X, calculate VIF and save in dataframe
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])] 
vif["features"] = X.columns
vif.round(1)
```

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
```

```python
X_test = ss.fit_transform(X_test)
```

Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

Predict Test Data

```python
lr.predict(X_test)
```

```python
y_pred = lr.predict(X_test)
```

Model Accuracy

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

Significant variables

```python
import statsmodels.api as sm
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
